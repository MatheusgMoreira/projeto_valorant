```python
import pandas as pd
```

```python
url = 'valorant_games.csv'
df = pd.read_csv(url)
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = df.loc[df['rank'] != 'Placement']
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df
```

```python
df_agents_stats = df.groupby('agent').agg(
    kills=('kills', 'sum'),
    deaths=('deaths', 'sum'),
    assists=('assists', 'sum'),
    round_wins=('round_wins', 'sum'),
    round_losses=('round_losses', 'sum'),
    headshot_pct=('headshot_pct', 'mean'),
    kdr_avg=('kdr', 'mean'),
    avg_dmg=('avg_dmg', 'mean'),
    acs=('acs', 'mean'),
).reset_index()

df_agents_stats['headshot_pct'] = df_agents_stats['headshot_pct'].fillna(0).astype(int)

# Adicionando a coluna 'role'
role_mapping = {
    'Astra': 'Controlador',
    'Breach': 'Iniciador',
    'Brimstone': 'Controlador',
    'Cypher': 'Sentinela',
    'Fade': 'Iniciador',
    'KAY/O': 'Iniciador',
    'Killjoy': 'Sentinela',
    'Omen': 'Controlador',
    'Phoenix': 'Duelista',
    'Viper': 'Controlador',
    'Vyse': 'Sentinela'
}

# Adicionando a coluna 'role'
agent_icon_mapping = {
    'Astra': 'https://static.wikia.nocookie.net/valorant/images/e/e0/Astra_Artwork_Full.png/revision/latest?cb=20220810202708',
    'Breach': 'https://static.wikia.nocookie.net/valorant/images/2/24/Breach_Artwork_Full.png/revision/latest?cb=20220810202713',
    'Brimstone': 'https://static.wikia.nocookie.net/valorant/images/8/81/Brimstone_Artwork_Full.png/revision/latest?cb=20220810202719',
    'Cypher': 'https://static.wikia.nocookie.net/valorant/images/5/55/Cypher_Artwork_Full.png/revision/latest?cb=20220810202731',
    'Fade': 'https://static.wikia.nocookie.net/valorant/images/e/e8/Fade_Artwork_Full.png/revision/latest?cb=20220810202738',
    'KAY/O': 'https://static.wikia.nocookie.net/valorant/images/5/57/KAYO_Artwork_Full.png/revision/latest?cb=20220810202747',
    'Killjoy': 'https://static.wikia.nocookie.net/valorant/images/8/81/Killjoy_Artwork_Full.png/revision/latest?cb=20220810202751',
    'Omen': 'https://static.wikia.nocookie.net/valorant/images/0/0e/Omen_Artwork_Full.png/revision/latest?cb=20220810202807',
    'Phoenix': 'https://static.wikia.nocookie.net/valorant/images/9/90/Phoenix_Artwork_Full.png/revision/latest?cb=20220810202811',
    'Viper': 'https://static.wikia.nocookie.net/valorant/images/8/85/Viper_Artwork_Full.png/revision/latest?cb=20220810202837',
    'Vyse': 'https://static.wikia.nocookie.net/valorant/images/d/d4/Vyse_Artwork_Full.png/revision/latest?cb=20240827165746'
}

df_agents_stats['role'] = df_agents_stats['agent'].map(role_mapping)
df_agents_stats['agent_icon'] = df_agents_stats['agent'].map(agent_icon_mapping)

# Exibindo o DataFrame atualizado
df_agents_stats
```

```python
df_maps_stats = df.groupby('map').agg(
    kills=('kills', 'sum'),
    kills_avg=('kills', 'mean'),
    deaths=('deaths', 'sum'),
    deaths_avg=('deaths', 'mean'),
    assists=('assists', 'sum'),
    assists_avg=('assists', 'mean'),
    kdr_avg=('kdr', 'mean'),
    acs=('acs', 'mean'),
    victories=('outcome', lambda x: (x == 'Win').sum()),  # Contando vitórias
    losses=('outcome', lambda x: (x == 'Loss').sum())     # Contando derrotas
).reset_index()

# Convertendo as médias para inteiros
df_maps_stats[['kills_avg', 'deaths_avg', 'assists_avg']] = df_maps_stats[['kills_avg', 'deaths_avg', 'assists_avg']].astype(int)

# Adicionando o agente mais jogado de cada mapa
most_played_agent = (
    df.groupby('map')['agent']
    .apply(lambda x: x.value_counts().idxmax())  # Identificando o agente mais frequente por mapa
    .reset_index(name='most_played_agent')  # Transformando em DataFrame com nova coluna
)

# Combinando os dados do agente mais jogado ao DataFrame principal
df_maps_stats = pd.merge(df_maps_stats, most_played_agent, on='map')

map_icon_mapping = {
    'Ascent': 'https://static.wikia.nocookie.net/valorant/images/e/e7/Loading_Screen_Ascent.png/revision/latest?cb=20200607180020',
    'Lotus': 'https://static.wikia.nocookie.net/valorant/images/d/d0/Loading_Screen_Lotus.png/revision/latest?cb=20230106163526',
    'Bind': 'https://static.wikia.nocookie.net/valorant/images/2/23/Loading_Screen_Bind.png/revision/latest?cb=20200620202316',
    'Haven': 'https://static.wikia.nocookie.net/valorant/images/7/70/Loading_Screen_Haven.png/revision/latest?cb=20200620202335',
    'Split': 'https://static.wikia.nocookie.net/valorant/images/d/d6/Loading_Screen_Split.png/revision/latest?cb=20230411161807',
    'Sunset': 'https://static.wikia.nocookie.net/valorant/images/5/5c/Loading_Screen_Sunset.png/revision/latest?cb=20230829125442',
    'Pearl': 'https://static.wikia.nocookie.net/valorant/images/a/af/Loading_Screen_Pearl.png/revision/latest?cb=20220622132842',
    'Icebox': 'https://static.wikia.nocookie.net/valorant/images/1/13/Loading_Screen_Icebox.png/revision/latest?cb=20201015084446',
    'Fracture': 'https://static.wikia.nocookie.net/valorant/images/f/fc/Loading_Screen_Fracture.png/revision/latest?cb=20210908143656',
    'Breeze': 'https://static.wikia.nocookie.net/valorant/images/1/10/Loading_Screen_Breeze.png/revision/latest?cb=20210427160616',
    'Abyss': 'https://static.wikia.nocookie.net/valorant/images/6/61/Loading_Screen_Abyss.png/revision/latest?cb=20240730145619'
}

df_maps_stats['map_icon'] = df_maps_stats['map'].map(map_icon_mapping)

df_maps_stats
```

```python
df_rank_stats = df.groupby('rank').agg(
    kills=('kills', 'sum'),
    deaths=('deaths', 'sum'),
    assists=('assists', 'sum'),
    round_wins=('round_wins', 'sum'),
    round_losses=('round_losses', 'sum'),
    headshot_pct=('headshot_pct', 'mean'),
    kdr_avg=('kdr', 'mean')).reset_index()

df_rank_stats['headshot_pct'] = df_rank_stats['headshot_pct'].fillna(0).astype(int)

rank_icon_mapping = {
    'Silver 2': 'https://static.wikia.nocookie.net/valorant/images/e/e9/Silver_2_Rank.png/revision/latest?cb=20200623203410',
    'Silver 3': 'https://static.wikia.nocookie.net/valorant/images/d/d7/Silver_3_Rank.png/revision/latest?cb=20200623203413',
    'Gold 1': 'https://static.wikia.nocookie.net/valorant/images/6/65/Gold_1_Rank.png/revision/latest?cb=20200623203413',
    'Gold 2': 'https://static.wikia.nocookie.net/valorant/images/0/02/Gold_2_Rank.png/revision/latest?cb=20200623203415',
    'Gold 3': 'https://static.wikia.nocookie.net/valorant/images/2/27/Gold_3_Rank.png/revision/latest?cb=20200623203417',
    'Platinum 1': 'https://static.wikia.nocookie.net/valorant/images/9/96/Platinum_1_Rank.png/revision/latest?cb=20200623203419',
    'Platinum 2': 'https://static.wikia.nocookie.net/valorant/images/5/5a/Platinum_2_Rank.png/revision/latest?cb=20200623203606',
    'Platinum 3': 'https://static.wikia.nocookie.net/valorant/images/1/1b/Platinum_3_Rank.png/revision/latest?cb=20200623203607',
    'Diamond 1': 'https://static.wikia.nocookie.net/valorant/images/a/ae/Diamond_1_Rank.png/revision/latest?cb=20200623203609',
    'Diamond 2': 'https://static.wikia.nocookie.net/valorant/images/6/6a/Diamond_2_Rank.png/revision/latest?cb=20200623203610',
    'Diamond 3': 'https://static.wikia.nocookie.net/valorant/images/6/6a/Diamond_2_Rank.png/revision/latest?cb=20200623203610',
    'Ascendant 1': 'https://static.wikia.nocookie.net/valorant/images/e/e5/Ascendant_1_Rank.png/revision/latest?cb=20220616175506',
    'Ascendant 2': 'https://static.wikia.nocookie.net/valorant/images/1/1e/Ascendant_2_Rank.png/revision/latest?cb=20220616175514'
}

df_rank_stats['rank_icon'] = df_rank_stats['rank'].map(rank_icon_mapping)

df_rank_stats
```

```python
def most_played(group, column):
    return group[column].value_counts().idxmax()

df_grouped_stats_rma = (
    df.groupby('rank')
    .apply(lambda group: pd.Series({
        'most_played_map': most_played(group, 'map'),
        'most_played_agent': most_played(group, 'agent'),
        'agent_kills': group.loc[group['agent'] == most_played(group, 'agent'), 'kills'].sum(),
        'agent_deaths': group.loc[group['agent'] == most_played(group, 'agent'), 'deaths'].sum(),
        'agent_assists': group.loc[group['agent'] == most_played(group, 'agent'), 'assists'].sum(),
    }))
    .reset_index()
)

df_grouped_stats_rma
```

```python
df_rank_by_date = (
    df.groupby(['rank', 'year', 'month'])
    .agg(
        total_kills=('kills', 'sum'),
        total_deaths=('deaths', 'sum'),
        total_assists=('assists', 'sum'),
        games_played=('rank', 'count')
    )
    .reset_index()
)

df_rank_by_date = df_rank_by_date.sort_values(by=['year', 'month', 'rank']).reset_index(drop=True)
df_rank_by_date
```

```python
file_path = 'dataframe_valorant.xlsx'
```

```python
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    df_agents_stats.to_excel(writer, sheet_name='Agents Stats', index=False)
    df_maps_stats.to_excel(writer, sheet_name='Maps Stats', index=False)
    df_rank_stats.to_excel(writer, sheet_name='Rank Stats', index=False)
    df_grouped_stats_rma.to_excel(writer, sheet_name='Grouped Stats Rank Map Agent', index=False)
    df_rank_by_date.to_excel(writer, sheet_name='Rank by Date', index=False)
```
